```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
from datetime import timedelta
```

```python
def Plotbox(mdtf1,mcolum,morder):
    plt.rcParams['figure.dpi'] = 140
    sns.boxplot(x='test', y=mcolum, data=mdtf1, order=morder)
    ax = plt.gca()
    ax.grid(True)
    plt.show()
    
def Append8(mdtf1,mdtf2,mdtf3,mdtf4,mdtf5,mdtf6,mdtf7,mdtf8):
    mdtf1 = mdtf1.append(mdtf2)
    mdtf1 = mdtf1.append(mdtf3)
    mdtf1 = mdtf1.append(mdtf4)
    mdtf1 = mdtf1.append(mdtf5)
    mdtf1 = mdtf1.append(mdtf6)
    mdtf1 = mdtf1.append(mdtf7)
    return mdtf1.append(mdtf8)

def Append8andCast(mdtf1,mdtf2,mdtf3,mdtf4,mdtf5,mdtf6,mdtf7,mdtf8, O_column, simbol,value):
    mdtf1 = Append8(mdtf1,mdtf2,mdtf3,mdtf4,mdtf5,mdtf6,mdtf7,mdtf8)
    mdtf1[O_column] = CastOutLayers(mdtf1, O_column, simbol,value)
    return mdtf1

def getDates3(dft):
    sal = pd.to_datetime(dft.dateStore, format="%Y-%m-%d %H:%M:%S.%f")
    dft['latency in seconds'] = (sal-pd.to_datetime(dft.Agre_date, format="%Y-%m-%d %H:%M:%S.%f"))/ np.timedelta64(1, 's')
    return dft[['latency in seconds','test']]
    
def AppendAndPlot(df1,df2,df3,df4,df5,df6,
                  O_column, simbol, value, morder):
    ndf = Append(df1, df2, df3, df4, df5, df6).reset_index(drop=True)
    ndf[O_column] = CastOutLayers(ndf, O_column, simbol,value)
    Plotbox(ndf,O_column,morder)

def CastOutLayers(mdtf1, column, lowerorgreater,mlimit):
    aux = []
    for row in mdtf1[column]:
        if type(row) is str:
            row =row.replace(".", "")
        if lowerorgreater == "<":
            if float(row) > mlimit:
                aux.append(mlimit)
            else:
                aux.append(float(row))
        elif lowerorgreater == ">":
            if float(row) < mlimit:
                aux.append(mlimit)
            else:
                aux.append(float(row))
    return aux
```

```python
def get_latency(dft):
    t_llegada = pd.to_datetime(dft.llegada, format="%Y-%m-%d %H:%M:%S.%f")
    t_envio = pd.to_datetime(dft.salida,  format="%Y-%m-%d %H:%M:%S.%f")
    dft['latency in seconds'] = ((t_llegada - t_envio)/2)/ np.timedelta64(1, 's')
    return dft[['latency in seconds','test']]

def Get_CSV(nameCSV):
    return pd.read_csv(nameCSV, names = ["timestamp1", "salida","llegada"])

def getDates3(dft):
    sal = pd.to_datetime(dft.Hora, format="%Y-%m-%d %H:%M:%S.%f")
    dft['latency in seconds'] = (sal-pd.to_datetime(dft.timestamp2, format="%Y-%m-%d %H:%M:%S.%f"))/ np.timedelta64(1, 's')
    return dft[['latency in seconds','test']]

def CastAndPlot2(df1,df2,N_column,lowerorgreater,mlimit,morder,leyenda):
    plt.rcParams['figure.dpi'] = 200
    plt.plot(morder,df1)
    plt.plot(morder,df2)
    ax = plt.gca()
    ax.grid(True)
    ax.grid(color='black', linestyle='-', linewidth=0.05)
    ax.set_axis_bgcolor('white')
    plt.legend(leyenda, loc='upper center')
    plt.xticks(np.arange(700, 825, step=25))
    plt.xlabel('NÂº Alarmas')
    plt.ylabel(N_column)
    plt.savefig(N_column+'_O.eps', format='eps', dpi=1000)
    
def getMean(df1,df2,df3,df4,df5,N_column):
    return [df1[N_column].mean(),df2[N_column].mean(),df3[N_column].mean(),df4[N_column].mean(),df5[N_column].mean()]
```

```python
dft100 = Get_CSV('02_19/latencyRaspberry/wifi100.csv')
dft100  ["test"]= "100"
lat100 = get_latency(dft100)

dft200 = Get_CSV('02_19/latencyRaspberry/wifi200.csv')
dft200  ["test"]= "200"
lat200 = get_latency(dft200)

dft300 = Get_CSV('02_19/latencyRaspberry/wifi300.csv')
dft300  ["test"]= "300"
lat300 = get_latency(dft300)

dft400 = Get_CSV('02_19/latencyRaspberry/wifi400.csv')
dft400  ["test"]= "400"
lat400 = get_latency(dft400)

dft500 = Get_CSV('02_19/latencyRaspberry/wifi500.csv')
dft500  ["test"]= "500"
lat500 = get_latency(dft500)

dft600 = Get_CSV('02_19/latencyRaspberry/wifi600.csv')
dft600  ["test"]= "600"
lat600 = get_latency(dft600)

dft700 = Get_CSV('02_19/latencyRaspberry/wifi700.csv')
dft700  ["test"]= "700"
lat700 = get_latency(dft700)

dft800 = Get_CSV('02_19/latencyRaspberry/wifi800.csv')
dft800  ["test"]= "800"
lat800 = get_latency(dft800)
```
